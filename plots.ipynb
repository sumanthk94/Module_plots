```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('All-India-Rainfall-Act_Dep_1901_to_2019_0.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
plt.rcParams.update({'font.size':20})
sns.pairplot(data[['JUN','JUL','AUG','SEP','JUN-SEP']],diag_kind=None)
# plt.title('Correlation of rainfall (MM) between different months' )
plt.savefig('correlation of rainfall data.jpg',dpi=1200,
          orientation='portrait',facecolor='w',edgecolor = 'w');
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=data[['JUN','JUL','AUG','SEP']])
plt.rcParams.update({'font.size':20})
plt.xlabel('Months',fontsize=20)
plt.ylabel('Rainfall data in MM',fontsize=20)
plt.grid()
plt.savefig('boxplot of rainfall data.jpg',dpi=1200)
```

```python
x = data['YEAR'].values
y = data[['JUN','JUL','AUG','SEP']].values
```

```python
fig, ax = plt.subplots(figsize=(20,12))
ax.plot(x,data['JUN'],label='June')
ax.plot(x,data['JUL'],label='July')
ax.plot(x,data['AUG'],label='August')
ax.plot(x,data['SEP'],label='September')
plt.rcParams.update({'font.size':20})
plt.xlabel('Years',fontsize=20)
plt.ylabel('Rainfall in MM',fontsize=20)
ax.legend(loc='upper right', frameon=False)
plt.savefig('Rainfall Lineplot.jpg',dpi=1200)
```
